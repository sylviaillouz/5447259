At least half of the contigs/scaffolds in the assembly will have the length of the N50 or greater. The mean or median contig length could be skewed by a couple of very very long contigs and the rest extremely short contigs (say 3 extremely long contigigs and thousands of 5 base pair contigs could have the same average as all mid sized contigs).
